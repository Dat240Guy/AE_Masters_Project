CQUAD8 Sympy Calc

```python
import sympy as sp
import numpy as np

xi, eta = sp.symbols('xi eta')
display(xi, eta)
```

```python
N1 = 1/4*(1-xi)*(1-eta)
N2 = 1/4*(1+xi)*(1-eta)
N3 = 1/4*(1+xi)*(1+eta)
N4 = 1/4*(1-xi)*(1+eta)

Coords = np.array([[-1, -1],
                    [1, -1],
                    [1, 1],
                    [-1, 1]])

```

```python
Ns = [N1, N2, N3, N4]
for N in Ns:
    display(sp.simplify(N))
    calc = [N.subs({xi: c[0], eta: c[1]}) for c in Coords]
    print(calc)
    print()
```

dN_dxi

```python
for N in Ns:
    dN_dXi = sp.diff(N, xi)
    # display(sp.simplify(dN_dXi))
    display(sp.expand(dN_dXi))
```

dN_deta

```python
for N in Ns:
    dN_deta = sp.diff(N, eta)
    display(sp.simplify(dN_deta))
```
