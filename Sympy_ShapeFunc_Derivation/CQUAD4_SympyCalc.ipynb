CQUAD8 Sympy Calc

```python
import sympy as sp
import numpy as np

xi, eta = sp.symbols('xi eta')
display(xi, eta)
```

```python
N1 = 1/4*(1-xi)*(1-eta)
N2 = 1/4*(1+xi)*(1-eta)
N3 = 1/4*(1+xi)*(1+eta)
N4 = 1/4*(1-xi)*(1+eta)

Coords = np.array([[-1, -1],
                    [1, -1],
                    [1, 1],
                    [-1, 1]])

```

```python
Ns = [N1, N2, N3, N4]
for N in Ns:
    display(sp.simplify(N))
    calc = [N.subs({xi: c[0], eta: c[1]}) for c in Coords]
    print(calc)
    print()
```

dN_dxi

```python
dN_dXi_Values = []
xiSub = 1/sp.sqrt(3)
etaSub = 1/sp.sqrt(3)
for N in Ns:
    dN_dXi = sp.diff(N, xi)
    # display(sp.simplify(dN_dXi))
    display(sp.expand(dN_dXi))
    val = dN_dXi.subs({xi:xiSub, eta:etaSub}).evalf()
    dN_dXi_Values.append(val)
    display(val)
```

dN_deta

```python
dN_dEta_Values = []
xiSub = 1/sp.sqrt(3)
etaSub = 1/sp.sqrt(3)
for N in Ns:
    dN_deta = sp.diff(N, eta)
    display(sp.simplify(dN_deta))
    val = dN_deta.subs({xi:xiSub, eta:etaSub}).evalf()
    dN_dEta_Values.append(val)
    display(val)
```

```python
B1 = sp.Matrix([[1, 0, 0, 0],
                [0, 0, 0, 1], 
                [0, 1, 1, 0]])

Nx = dN_dXi_Values
Ne = dN_dEta_Values
B3 = sp.Matrix([[Nx[0], 0, Nx[1], 0, Nx[2], 0, Nx[3], 0],
               [Ne[0], 0, Ne[1], 0, Ne[2], 0, Ne[3], 0],
               [0, Nx[0], 0, Nx[1], 0, Nx[2], 0, Nx[3]],
               [0, Ne[0], 0, Ne[1], 0, Ne[2], 0, Ne[3]]])

display(B1 @ B3)
```
