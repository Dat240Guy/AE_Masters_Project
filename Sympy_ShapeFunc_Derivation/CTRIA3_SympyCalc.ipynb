```python
import sympy as sp
import numpy as np

xi, eta = sp.symbols('xi eta')
display(xi, eta)
```

```python
N1 = 1 - xi - eta
N2 = xi
N3 = eta


Coords = np.array([[0, 0],
                    [1, 0],
                    [0, 1]])
```

```python
Ns = [N1, N2, N3]
for N in Ns:
    display(sp.simplify(N))
    calc = [N.subs({xi: c[0], eta: c[1]}) for c in Coords]
    print(calc)
    print()
```

```python
dN_dXi_Values = []
xiSub = 1/sp.sqrt(3)
etaSub = 1/sp.sqrt(3)
for N in Ns:
    dN_dXi = sp.diff(N, xi)
    # display(sp.simplify(dN_dXi))
    display(sp.expand(dN_dXi))
    val = dN_dXi.subs({xi:xiSub, eta:etaSub}).evalf()
    dN_dXi_Values.append(val)
    display(val)
```

```python
dN_dEta_Values = []
xiSub = 1/sp.sqrt(3)
etaSub = 1/sp.sqrt(3)
for N in Ns:
    dN_deta = sp.diff(N, eta)
    display(sp.simplify(dN_deta))
    val = dN_deta.subs({xi:xiSub, eta:etaSub}).evalf()
    dN_dEta_Values.append(val)
    display(val)
```
