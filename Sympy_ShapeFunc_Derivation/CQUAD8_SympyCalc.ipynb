CQUAD8 Sympy Calc

```python
import sympy as sp
import numpy as np

xi, eta = sp.symbols('xi eta')
display(xi, eta)
```

```python
N1 = -1/4 * (1-eta) * (1-xi) * (1+xi+eta)
N2 = -1/4 * (1-eta) * (1+xi) * (1-xi+eta)
N3 = -1/4 * (1+eta) * (1+xi) * (1-xi-eta)
N4 = -1/4 * (1+eta) * (1-xi) * (1+xi-eta)
N5 =  1/2 * (1-eta) * (1-xi) * (1+xi)
N6 =  1/2 * (1-eta) * (1+xi) * (1+eta)
N7 =  1/2 * (1+eta) * (1-xi) * (1+xi)
N8 =  1/2 * (1-eta) * (1-xi) * (1+eta)

Coords = np.array([[-1, -1], #LL
                    [1, -1],  #LR
                    [1, 1],   #UR
                    [-1, 1],  #UL
                    [0, -1],  #MB
                    [1, 0],   #MR
                    [0, 1],   #MU
                    [-1, 0]]) #ML
```

```python
Ns = [N1, N2, N3, N4, N5, N6, N7, N8]
for N in Ns:
    display(sp.simplify(N))
    calc = [N.subs({xi: c[0], eta: c[1]}) for c in Coords]
    print(calc)
    print()
```

dN_dxi

```python
for N in Ns:
    dN_dXi = sp.diff(N, xi)
    display(sp.simplify(dN_dXi))
```

dN_deta

```python
for N in Ns:
    dN_deta = sp.diff(N, eta)
    display(sp.simplify(dN_deta))
```
